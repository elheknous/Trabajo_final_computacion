```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Leer archivo

xs, ys, us = [], [], []
with open("output_heat.txt") as f:
    for line in f:
        line = line.strip()
        if not line or line.startswith("#"):
            continue
        parts = line.split()
        if len(parts) == 3:
            x, y, u = map(float, parts)
        else:
            x, y, u = float(parts[2]), float(parts[3]), float(parts[4])
        xs.append(x); ys.append(y); us.append(u)

xs = np.array(xs)
ys = np.array(ys)
us = np.array(us)

# Reconstruir grilla
xv = np.unique(xs)
yv = np.unique(ys)
X, Y = np.meshgrid(xv, yv, indexing="ij")

U = np.zeros_like(X)
xi = {x:i for i,x in enumerate(xv)}
yi = {y:i for i,y in enumerate(yv)}
for x,y,u in zip(xs,ys,us):
    U[xi[x], yi[y]] = u
```

```python

# Grfico 3D
fig = plt.figure(figsize=(25,6))
ax = fig.add_subplot(111, projection="3d")

surf = ax.plot_surface(X, Y, U,
                        cmap="viridis",
                        linewidth=0,
                        antialiased=True)

ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("u(x,y,t)")
ax.set_title("Solucin 3D de la ecuacin del calor")

fig.colorbar(surf, shrink=0.5, aspect=10)
plt.show()
```
